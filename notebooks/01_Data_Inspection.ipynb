### Investigating Zero-Income Loans

Some loans have `annual_inc == 0`. This is worth checking separately because zero income may represent a data-quality issue or unusual borrower reporting, not simply a normal low-income segment.

```python
(df["annual_inc"] == 0).sum()
```

```python
zero_income = df[df["annual_inc"] == 0]
zero_income[
[
  "annual_inc",
  "loan_amnt",
  "grade",
  "home_ownership",
  "emp_length",
  "verification_status",
  "loan_status"
]]
```

```python
display(zero_income.groupby("verification_status")
.size()
.div(len(zero_income))
.mul(100)
.round(2)
)

#compare loan amt
display(zero_income["loan_amnt"].describe().round(2))

min_income = df[df["annual_inc"].between(1, 24999) ]
display(min_income["loan_amnt"].describe().round(2))
```

```python

#compare loan status with low income group
display(
  zero_income.groupby("loan_status")
  .size()
  .div(len(zero_income))
  .mul(100)
  .round(2)
  .to_frame("percentage")
)

display(
  min_income.groupby("loan_status")
  .size()
  .div(len(min_income))
  .mul(100)
  .round(2)
  .to_frame("percentage")
)

```

### Observation

There are 322 zero-income loans. Their loan amounts are surprisingly high compared with borrowers earning $1-$24,999: the zero-income median loan amount is $20,000 versus $7,200 for the low-income comparison group. Most zero-income rows are `Not Verified`, which supports treating this group cautiously as a possible reporting or data-quality issue.

The loan-status mix for zero-income borrowers is similar to the low-income comparison group, but the zero-income group is small and mostly still `Current`, so this does not prove the records are reliable.

### DTI Exploration


```python
# Basic DTI distribution
df["dti"].describe(percentiles=[0.01, 0.05, 0.25, 0.5, 0.75, 0.95, 0.99]).round(2)
```

```python
dti_summary = pd.Series({
  "missing": df["dti"].isna().sum(),
  "missing_pct": (df["dti"].isna().mean() * 100).round(2),
  "zero": (df["dti"] == 0).sum(),
  "over_100": (df["dti"] > 100).sum(),
  "equal_999": (df["dti"] == 999).sum()
})
dti_summary
```

```python
cols = [
    "dti",
    "annual_inc",
    "loan_amnt",
    "installment",
    "verification_status",
    "home_ownership",
    "purpose",
    "grade",
    "sub_grade",
    "loan_status"
    
]

df.loc[
  df["dti"] > 100,
  cols
].sort_values(by = "dti", ascending=False)
```

#### Observation

DTI missingness is low: 328 rows, about 0.25% of the dataset. The bigger issue is extreme values. The 99th percentile is 59.85, but the maximum is 999, with 394 rows above 100 and 13 rows exactly equal to 999. These values should be treated as extreme or special cases during preprocessing, for example by flagging high DTI and possibly capping the numeric value.

#### Non numerical column inspection

```python
cat_cols = df.select_dtypes(include=["object", "string"]).columns

cat_summary = pd.DataFrame({
  "dtype": df[cat_cols].dtypes,
  "missing": df[cat_cols].isna().sum(),
  "missing_pct": df[cat_cols].isna().mean() * 100,
  "unique": df[cat_cols].nunique(dropna=False)
}).sort_values("unique", ascending=False)

cat_summary
```

#### Observations

- 4 variables were identified as numeric values stored as strings.
- 5 variables can be converted to datetime format.
- Most categorical variables exhibit low cardinality (<50 categories) and are suitable for encoding.
- emp_title (44,610 unique values) has a very high-cardinality and will be evaluated separately.
- Several hardship and settlement related variables contain >99% missing values and appear to contain post-origination information. These variables are potential sources of target leakage and will likely be excluded from modeling.
